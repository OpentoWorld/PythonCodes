```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.bar([1.5,2.6,4.4],[40,55,89])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x=np.array(['Science','Maths','English','Social Studies'])
y=np.array([60,30,40,80])
plt.bar(x,y)
plt.ylabel('Marks of a student')
plt.xlabel('Subjects Name')
plt.grid
plt.savefig("Bar_Graph_representation_of_a_student.png")
plt.show()
```

```python
#Plotting a dictionary using a bar chart
dictionary={'A':25,'B':70,'C':55,'D':90}
for i,key in enumerate(dictionary):
    plt.bar(i,dictionary[key])
plt.xticks(np.arange(len(dictionary)),dictionary.keys())
plt.show()
```
