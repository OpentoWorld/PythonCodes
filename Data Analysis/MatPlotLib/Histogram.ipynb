```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y=np.random.randn(100,100)
plt.hist(y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.savefig("Plot1.png")
#plt.show()
```
