```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=np.arange(5)
print (n)
plt.plot(n,n,label='linear')
plt.plot(n,n*n,label='square')
plt.plot(n,n*n*n,label='cube')
plt.grid(True)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Polynomial Graph')
plt.legend()
plt.savefig('plot.png')
plt.show()
```
