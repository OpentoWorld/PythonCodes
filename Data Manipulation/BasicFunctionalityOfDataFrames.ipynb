```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x={'odd':np.arange(1,50,2),'even':np.arange(0,50,2)}
print (x['odd'])
print (x['even'])
xy=pd.DataFrame(x)
print ("Returns the sum of the value for the requested axes")
print (xy.sum())
print ("Returns the standard deviation of the value for the requested axes")
print (xy.std())
```

```python
df=pd.DataFrame(np.random.rand(5,4),columns=['col1','col2','col3','col4'])
print ("Iterates over each column as key,value")
for key,value in df.iteritems():
    print (key,value)
plt.plot(df)
plt.show()
```

```python
print ("Iterates over each row as key,value pair")
for key,value in df.iterrows():
    print (key,value)
print ("Returns an iteration yielding a named tuple for each row")
for row in df.itertuples():
    print (row)
```
