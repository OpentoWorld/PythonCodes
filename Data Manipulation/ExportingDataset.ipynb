```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Allcountries.csv")
print (dataset.shape)
```

```python
print (dataset.dtypes)#data types
```

```python
print ('Top 10 Countries list')
print (dataset.head(10))
print ('=========================================================================')
print ('Bottom 5 Countries list')
print (dataset.tail(5))
```

```python
#Statistical summary of datasets
print (dataset.describe())
```

```python
#Extra
selected_data=dataset.loc[:,['Country','LandArea']]
for i in selected_data.itertuples():
    if i[2]>2000:
        print (i)
```

```python
plt.figure(figsize=(10,10))
selected_data=dataset.loc[:,['Country','GDP','BirthRate']]
x=np.array(selected_data["GDP"])
y=np.array(selected_data["BirthRate"])
print (x,y)
plt.scatter(x,y)
plt.xlabel('GDP=====>')
plt.ylabel('BirthRate====>')
plt.xlim(0,2000)
plt.savefig("BirthRate_GDP.png")
plt.show()

```

```python
sorted_data=selected_data.sort_index(by='GDP',ascending=False)
selected_sorted_data=sorted_data.iloc[:10]
print (selected_sorted_data)
plt.pie(selected_sorted_data['GDP'],labels=selected_sorted_data['Country'])
plt.show()
```
